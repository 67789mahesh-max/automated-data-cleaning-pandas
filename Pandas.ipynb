```python
df.dropna(axis=0,subset=['Transaction_ID'], inplace=True )
```

```python
df['Customer_Name'].value_counts(dropna=False)
```

```python
df['Customer_Name'] = df['Customer_Name'].str.strip().str.title().fillna('N/A')
```

```python
df['Customer_Name'].value_counts(dropna=False)
```

```python
df['Email'].value_counts(dropna=False)
```

```python
df['Email']=df['Email'].fillna('N/A')
```

```python
df['Age'].dtype
```

```python
df['Age'].value_counts()
```

```python
df[df['Age'].str.startswith('-', na=True)]
```

```python
df[df['Age'].str.startswith('-', na=False)]
```

```python
df['Age'] = df['Age'].replace({'-': '', 'twenty': '20', np.nan: 'N/A','250':'N/A'}, regex=True).str.strip()
```

```python

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
df['Age'][df['Age'].isnull()]
```

```python
df['Age'].value_counts().to_frame()
```

```python
df['Join_Date'].isnull().sum()
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], format='mixed', errors='coerce')
```

```python
df['Product_Category'].value_counts(dropna=False)
```

```python
df['Product_Category'].value_counts(dropna=False)
```

```python
df['Product_Category'] = df['Product_Category'].str.title().replace({'Elctronics':'Electronics', '_':' & '},regex=True).str.strip().fillna('No Category').astype('category')
```

```python
df['Product_Category'].dtype
```

```python
df[df['Price'].isnull()| df['Price'].str.startswith('-')]
```

```python

df['Price'] = df['Price'].replace({'USD': '', 'Free': '0'}, regex=True)
```

```python
# Convert to numeric, turning any remaining errors into NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['Quantity'].value_counts(dropna=False)
```

```python
df['Quantity'] = df['Quantity'].abs()
```

```python
df[df['Quantity'].isin([-2])]
```

```python
df['Discount_Percent'].value_counts(dropna=False).to_frame()
```

```python
df['Discount_Percent'].dtype
```

```python
df['Discount_Percent'] = df['Discount_Percent'].abs().fillna(0)
```

```python
df['Discount_Percent'].dtype
```

```python
import numpy as np

df['Discount_Percent'] = np.where(
    df['Discount_Percent'].between(0.0, 1.0), 
    df['Discount_Percent'], 
    0
)
```

```python
df['Price']=df['Price'].fillna(0)
```

```python
df['Price'].value_counts(dropna=False)
```

```python
df['Quantity']=df['Quantity'].fillna(0)
```

```python
# Round to 0 decimal places, then convert to Int64 (capital 'I' supports NaNs)
df['Quantity'] = df['Quantity'].round(0).astype('Int64')
```

```python
df.info()
```

```python
df['Total_Amount'] = (df['Price'] * df['Quantity']) * (1 - df['Discount_Percent'])
```

```python
df['Shipping_State'].value_counts(dropna=False)
```

```python

state_map = {
    'CA': 'California',
    'TX': 'Texas',
    'NY': 'New York',
    'N.Y.': 'New York',
    'teXas': 'Texas'
}


df['Shipping_State'] = (df['Shipping_State']
                        .replace(state_map)
                        .str.strip()
                        .str.title()
                        .fillna('N/A'))


print(df['Shipping_State'].value_counts())
```

```python
df['Customer_Review_Score']=df['Customer_Review_Score'].fillna(0).abs()
```

```python
df['Customer_Review_Score'].value_counts(dropna=False)
```

```python
df['Is_Premium_Member'].value_counts(dropna=False)
```

```python
state_map = {
    'False': 'NO',
    'True': 'Yes',
    '0': 'NO',
    '1': 'Yes',
    'Y': 'Yes',
    'N':'NO'
}


df['Is_Premium_Member'] = (df['Is_Premium_Member']
                        .replace(state_map)
                        .str.strip()
                        .str.title()
                        .fillna('N/A'))


print(df['Is_Premium_Member'].value_counts())
```
